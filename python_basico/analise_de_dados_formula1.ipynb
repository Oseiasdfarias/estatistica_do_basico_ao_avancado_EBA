#### Corrigindo campo `milliseconds` do DataFrame results

```python
df_results.head(2)
```

```python
df_results["milliseconds"] = (df_results["milliseconds"]
                              .replace("\\N", np.NAN))
df_results["milliseconds"].fillna(0, inplace=True)
df_results["milliseconds"] = df_results["milliseconds"].astype(int)
```

```python
df_results["milliseconds"].mean()
```

```python
df_results["milliseconds"] = (
    df_results["milliseconds"]
    .replace(0, df_results["milliseconds"].mean()))
```

#### Pré-processamento dos dados do DataFrame Pit Stops & Results

```python
df_pit_stops.info()
```

```python
df_pit_stops["raceId"] = df_pit_stops["raceId"].astype(str)
df_pit_stops["driverId"] = df_pit_stops["driverId"].astype(str)
```

```python
df_results.info()
```

```python
df_results["resultId"] = df_results["resultId"].astype(str)
df_results["raceId"] = df_results["raceId"].astype(str)
df_results["driverId"] = df_results["driverId"].astype(str)
```

#### Juntar DataFrame Pit Stops & Results

```python
results_pit_stops = df_results.merge(
    df_pit_stops, on=["raceId", "driverId"])
```

```python
results_pit_stops.head(3)
```

#### **Histograma do tempo de Pit Stops em Segundos**

```python
results_pit_stops["pit_seconds"] = results_pit_stops["milliseconds_x"]/1000
```

```python
results_pit_stops.columns
```

```python
fig = px.histogram(results_pit_stops, x="pit_seconds", nbins=10)
fig.show()
```

#### Verificando a relação entre o xxx e xxx usando o gráfico Scatter

```python
fig = px.scatter(results_pit_stops,
                 x="pit_seconds",
                 y="positionOrder")
fig.show()
```

```python
results_pit_stops = results_pit_stops.replace(["\\N", "R"], np.NAN)
results_pit_stops = results_pit_stops.dropna()
results_pit_stops
```

```python
lista_colunas = ['number', 'grid', 'position', 'positionText',
                 'positionOrder', 'points', 'laps', 'milliseconds_x',
                 'fastestLap', 'rank', 'milliseconds_y', 'pit_seconds']
(results_pit_stops[lista_colunas]
 .corr()
 .style.background_gradient(axis=0))
```
